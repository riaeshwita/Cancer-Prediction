```python
#Mapping[nominal values]
df['diagnosis'] = df['diagnosis'].map({'M':1,'B':0})
df.diagnosis
```

```python
#X contains all columns other than target, so drop diagnosis 
X = df.drop('diagnosis', axis=1)
X
```

```python
#Y is target: Target value here is diagnosis
Y = df['diagnosis']
Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
#multiple columns with different units of measurement, use standardization to improve performance
#Normalizzation: range of 0s and 1s: method: MinMaxScaler
```

```python
#Standardization
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
X_train = ss.fit_transform(X_train)
X_test = ss.fit_transform(X_test)
print(X_train)
print(X_test)
```

```python
#Logistic Regression

from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(X_train, Y_train)
```

```python
y_pred = lr.predict(X_test)
y_pred
```

```python
Y_test
```

```python
#higher accuracy score, better results 

from sklearn.metrics import accuracy_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns #used for visualization

print(accuracy_score(Y_test, y_pred))
cm = confusion_matrix(Y_test, y_pred)
print(cm)

sns.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Prediction')
plt.ylabel('True Values')
```

```python
lr_acc = accuracy_score(Y_test, y_pred)
```

```python
results = pd.DataFrame()
results
```

```python
tempresults = pd.DataFrame({'Algorithm': ['Logistic Regression Method'], 'Accuracy': [lr_acc]})
results = pd.concat([results, tempresults])
results[['Algorithm','Accuracy']]
```

```python
import matplotlib.pyplot as plt

feature1="texture_mean"
feature2="radius_mean"
# Assuming you have a binary target variable ("M" for malignant, "B" for benign)
target_variable = "diagnosis"

# Separate data for each class based on the target variable
class_0_data = df[df[target_variable] == "B"]
class_1_data = df[df[target_variable] == "M"]

# Extract the feature values for each class
feature1_class_0 = class_0_data[feature1]
feature2_class_0 = class_0_data[feature2]

feature1_class_1 = class_1_data[feature1]
feature2_class_1 = class_1_data[feature2]

# Create a scatter plot
plt.figure(figsize=(15, 10))

# Scatter points for class 0 (benign)
plt.scatter(feature1_class_0, feature2_class_0, c='b', label='Benign', marker='o')

# Scatter points for class 1 (malignant)
plt.scatter(feature1_class_1, feature2_class_1, c='r', label='Malignant', marker='x')

# Set labels and title
plt.xlabel(feature1)
plt.ylabel(feature2)
plt.title(f'Scatter Plot of {feature1} vs {feature2}')

# Add a legend
plt.legend(loc='upper right')

# Show the plot
plt.show()
```
